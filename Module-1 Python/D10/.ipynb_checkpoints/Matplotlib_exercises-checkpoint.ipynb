### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
plt.plot(data['month_number'],data['total_profit'])
```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
# Write your code here
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(data['month_number'],data['total_profit'],color = 'red', linewidth=3.0, linestyle='--',marker = 'o', mfc = 'black')
axes.set_xlabel('Month Number') 
axes.set_ylabel('Sold units number')
axes.set_title('Company Sales Data of Last Year')
axes.legend("Profit data of last year", loc=4)
```

### Exercise 3: Read all product sales data and show it using a multiline plot.
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

The graph should look like this:


```python
data.head()
```

```python
# Write your code here
fig, ax = plt.subplots(figsize=(10,8))

ax.plot(data['month_number'],data['facecream'],color = 'r', linewidth=3.0, linestyle='-',marker = 'o')
ax.plot(data['month_number'],data['facewash'],color = 'b', linewidth=3.0, linestyle='-',marker = 'o')
ax.plot(data['month_number'],data['toothpaste'],color = 'g', linewidth=3.0, linestyle='-',marker = 'o')
ax.plot(data['month_number'],data['bathingsoap'],color = 'c', linewidth=3.0, linestyle='-',marker = 'o')
ax.plot(data['month_number'],data['shampoo'],color = 'm', linewidth=3.0, linestyle='-',marker = 'o')
ax.plot(data['month_number'],data['moisturizer'],color = 'y', linewidth=3.0, linestyle='-',marker = 'o')

ax.set_xlabel('Month Number') 
ax.set_ylabel('Sold units number')
ax.set_title('Sales Data')
ax.legend()

```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.
Also, add a grid in the plot (the grid line style should be âââ).

The scatter plot should look like this:


```python
# Write your code here
fig = plt.figure()

hx = fig.add_axes([0.1,0.1,0.8,0.8])

hx.scatter(data['month_number'],data['toothpaste'])
hx.grid(linestyle='--')

hx.set_xlabel('Month Number') 
hx.set_ylabel('Sold units number')
hx.set_title('Tooth Paste Sales Data')
hx.legend()
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
# Write your code here
import numpy as np

fig, bx = plt.subplots()
ind = np.arange(len(data['month_number']))
width = 0.25


bx.bar(ind-width/2, data['facecream'], width, label ='Facecream Sales')
bx.bar(ind+width/2, data['facewash'], width, label ='Facewash Sales')

bx.set_xlabel('Month Number') 
bx.set_ylabel('Sales Units in Number')
bx.set_title('Facewash and Facecream Sales Data')
bx.legend()
bx.grid(linestyle='--')

fig.tight_layout()

plt.show()
```

### Exercise 6: Read the sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk.

```python
# Write your code here
fig, bx = plt.subplots()
ind = np.arange(len(data['month_number']))
width = 0.3


bx.bar(ind, data['bathingsoap'], width, label ='Bathingsoap Sales')

bx.set_xlabel('Month Number') 
bx.set_ylabel('Sales Units in Number')
bx.set_title('Bathing Soap Sales Data')
bx.legend()
bx.grid(linestyle='--')

fig.tight_layout()

plt.show()
fig.savefig("Ex6.png")
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
```

```python
face_cream = data['facecream'].sum()/data['total_profit'].sum()*100
face_wash = data['facewash'].sum()/data['total_profit'].sum()*100
tooth_paste = data['toothpaste'].sum()/data['total_profit'].sum()*100
bathing_soap = data['bathingsoap'].sum()/data['total_profit'].sum()*100
shampoo = data['shampoo'].sum()/data['total_profit'].sum()*100
moisturizer = data['moisturizer'].sum()/data['total_profit'].sum()*100

pie_data =[face_cream, face_wash,tooth_paste,bathing_soap,shampoo,moisturizer]

plt.pie(pie_data, labels = ['Facecream','Facewash','Toothpaste','Bathing Soap','Shampoo','Moisturizer'])
plt.show()
```
